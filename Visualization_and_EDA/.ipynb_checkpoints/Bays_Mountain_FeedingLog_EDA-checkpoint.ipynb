```python
# List of columns to retain
feeding_log_cols = [
    'Timestamp', 'Animal', 'Staff', 'Log Type', 'Date', 'Time',
    'Did you feed all individuals? ', 'If not, who did you feed?',
    'What did you feed? (Weight per animal) [<1/4 lb]',
    'What did you feed? (Weight per animal) [1/4 lb]',
    'What did you feed? (Weight per animal) [1/2 lb]',
    'What did you feed? (Weight per animal) [3/4 lb]',
    'What did you feed? (Weight per animal) [1 lb]',
    'What did you feed? (Weight per animal) [1.25 lb]',
    'What did you feed? (Weight per animal) [1.5 lb]',
    'What did you feed? (Weight per animal) [1.75 lb]',
    'What did you feed? (Weight per animal) [2 lb]',
    'What did you feed? (Weight per animal) [5 lb]',
    'What did you feed? (Weight per animal) [10 lb]',
    'What did you feed? (Weight per animal) [15 lb]',
    'What did you feed? (Weight per animal) [20 lb]',
    'If other', 'Deer feed scoops', 'Observation Type', 'Finding'
]
```

```python
filtered_df = dataset[dataset['Log Type'] == 'Feeding/Observation']
```

```python
feeding_data = filtered_df[feeding_log_cols]
```

```python
feeding_df = pd.DataFrame(feeding_data)
```

```python
feeding_df.rename(columns={'What did you feed? (Weight per animal) [<1/4 lb]':'0.15',
    'What did you feed? (Weight per animal) [1/4 lb]':'0.25',
    'What did you feed? (Weight per animal) [1/2 lb]':'0.5',
    'What did you feed? (Weight per animal) [3/4 lb]':'0.75',
    'What did you feed? (Weight per animal) [1 lb]':'1',
    'What did you feed? (Weight per animal) [1.25 lb]':'1.25',
    'What did you feed? (Weight per animal) [1.5 lb]':'1.5',
    'What did you feed? (Weight per animal) [1.75 lb]':'1.75',
    'What did you feed? (Weight per animal) [2 lb]':'2',
    'What did you feed? (Weight per animal) [5 lb]':'5',
    'What did you feed? (Weight per animal) [10 lb]':'10',
    'What did you feed? (Weight per animal) [15 lb]':'15',
    'What did you feed? (Weight per animal) [20 lb]':'20'}, inplace=True)
```

```python
feeding_df.head()
```

```python
feeding_df_split = feeding_df.copy()
```

```python
#splitting the values separated by comma into individual columns

columns_to_split = ['0.15','0.25','0.5','0.75','1','1.25','1.5','1.75','2','5','10','15','20']

for col in columns_to_split:
    if feeding_df_split[col].astype(str).str.contains(',').any():
        # Split and expand into new columns
        split_cols = feeding_df_split[col].str.split(',', expand=True)
        
        # Strip leading/trailing spaces from split values
        split_cols = split_cols.applymap(lambda x: x.strip() if isinstance(x, str) else x)
        
        # Rename new columns to indicate the origin
        split_cols.columns = [f"{col}_{i+1}" for i in range(split_cols.shape[1])]
        
        # Concatenate the new columns with the original DataFrame
        feeding_df_split = pd.concat([feeding_df_split.drop(columns=[col]), split_cols], axis=1)
    else:
        feeding_df_split = pd.concat([feeding_df_split.drop(columns=[col]), feeding_df_split[col]], axis=1)

print(feeding_df_split.head())
    
```

```python
feeding_df_other_col = feeding_df_split.copy()
```

```python
#This step is taken because there are many fields with empty weight to foodtype cells and the 'If other' field is popualted with the value
#So we are using the value from the 'If other' field to populate the food type in the weight columns before transposing them.

import re

weight_columns = ['0.15', '0.25', '0.5', '0.75', '1', '1.25', '1.5', '1.75','2', '5', '10','15', '20']
column_names = ['0.15_1', '0.25_1','0.5_1', '0.75', '1_1', '1.25', '1.5', '1.75','2_1', '5_1', '10_1','15', '20_1']
food_type_mapping = {
    'NB': 'Nebraska Brand',
    'nb': 'Nebraska Brand',
    'chicken': 'Chicken',
    'Chicken': 'Chicken',
    'veg': 'Fresh Vegetables',
    'Veg': 'Fresh Vegetables',
    'fruits': 'Fresh Fruits'
}

def extract_and_assign(row):
    if_other = str(row.get('If other', '')).lower()  # Convert to string and lowercase
    if not if_other or if_other == 'nan':
        return row
        
    matches = re.findall(r'(\d+\.\d+)\s*lb\s+of\s+(\w+)', if_other)
    for weight_str, food_code in matches:
        weight = float(weight_str)
        # Find the closest weight column
        closest_col = min(zip(weight_columns, column_names), key=lambda x: abs(float(x[0]) - float(weight)))[1]
        # Check if the cell is empty
        if pd.isna(row[closest_col]):
            # Map the food type
            food_type = food_type_mapping.get(food_code.lower(), 'OTHER')
            #row[closest_col] = f"{weight}lb - {food_type}"
            row[closest_col] = food_type
    return row

# Apply the function to each row
feeding_df_other_col = feeding_df_other_col.apply(extract_and_assign, axis=1)

# Save the updated DataFrame
#feeding_df_other_col.to_csv(r'C:\Users\Shauhardha\Downloads\Project\updated_dataset.csv', index=False)
```

```python
feeding_df_transpose = feeding_df_other_col.copy()

feeding_df_transpose.head()
```

```python
feeding_df_transpose.columns
```

```python
# Identify id columns and value columns
id_cols = ['Timestamp', 'Animal', 'Staff', 'Log Type', 'Date', 'Time',
           'Did you feed all individuals? ', 'If not, who did you feed?',
           'If other', 'Deer feed scoops', 'Observation Type', 'Finding']

value_cols = [col for col in feeding_df_transpose.columns if col not in id_cols]

melted_df_test = feeding_df_transpose.melt(
    id_vars=id_cols,
    value_vars=value_cols,
    var_name='Weight',
    value_name='Food Type'
)

melted_df_test['Weight'] = melted_df_test['Weight'].str.extract(r'(\d+\.\d+|\d+)').squeeze()


# Remove rows where Food Type is NaN
melted_df_test = melted_df_test.dropna(subset=['Food Type'])
melted_df_test = melted_df_test[melted_df_test['Food Type'] != '']
```

```python
print(melted_df_test[id_cols].isnull().sum())
```

```python
melted_df_test['If not, who did you feed?'] = melted_df_test['If not, who did you feed?'].fillna('All')
melted_df_test['If other'] = melted_df_test['If other'].fillna('None')
melted_df_test['Deer feed scoops'] = melted_df_test['Deer feed scoops'].fillna('0')
```

```python
melted_df_test['Weight'] = pd.to_numeric(melted_df_test['Weight'], errors='coerce')
```

```python
# Step 3: Aggregate by summing weights for the same food type
#aggregated_df_t = melted_df_test.groupby(id_cols + ['Food Type'], as_index=False)['Weight'].sum()


# Pivot the table
pivot_df_t = melted_df_test.pivot_table(
    index=id_cols,
    columns='Food Type',
    values='Weight',
    aggfunc='sum',
    fill_value=0
).reset_index()

# Flatten column names
pivot_df_t.columns.name = None
```

```python
pivot_df_t.head()

print(pivot_df_t.shape)

# Save the updated DataFrame
#pivot_df_t.to_csv(r'C:\Users\Shauhardha\Downloads\Project\pivoted_dataset.csv', index=False)
```

```python
pivot_df_t['Total_food_quantity'] = pivot_df_t[['Chicken', 'Fish', 'Fresh Fruit', 'Fresh Vegetables', 'Mazuri Omnivore', 'Nebraska Brand', 'OTHER', 'Whole prey']].sum(axis=1)

print(pivot_df_t['Total_food_quantity'])
```

## Visualization of Feeding data

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Sum the amounts for each food type
food_totals = feeding_log_viz[['Chicken', 'Fish', 'Fresh Fruit', 'Fresh Vegetables', 'Mazuri Omnivore', 'Nebraska Brand', 'OTHER', 'Whole prey']].sum()

# Plot
plt.figure(figsize=(10, 6)) 
food_totals.plot(kind='bar', color='skyblue')
plt.title('Total Food Fed by Type')
plt.xlabel('Food Type')
plt.ylabel('Total Amount Fed (lbs)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by Animal and sum the food amounts
animal_food_totals = feeding_log_viz.groupby('Animal')[['Chicken', 'Fish', 'Fresh Fruit', 'Fresh Vegetables', 'Mazuri Omnivore', 'Nebraska Brand', 'OTHER', 'Whole prey']].sum()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(animal_food_totals, annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('Feeding Patterns by Animal and Food Type')
plt.xlabel('Food Type')
plt.ylabel('Animal')
plt.show()
```

```python
# Sum the amounts for each food type
food_totals = feeding_log_viz[['Chicken', 'Fish', 'Fresh Fruit', 'Fresh Vegetables', 'Mazuri Omnivore', 'Nebraska Brand', 'OTHER', 'Whole prey']].sum()

# Plot
plt.figure(figsize=(8, 8))
plt.pie(food_totals, labels=food_totals.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Proportion of Food Types in Diet')
plt.show()
```

```python
# Plot
plt.figure(figsize=(8, 6))
plt.scatter(feeding_log_viz['Chicken'], feeding_log_viz['Whole prey'], color='purple', alpha=0.6)
plt.title('Relationship Between Nebraska Brand and Whole Prey')
plt.xlabel('Nebraska Brand (lbs)')
plt.ylabel('Whole Prey (lbs)')
plt.grid(True)
plt.show()
```

```python
# Convert 'Timestamp' to datetime
feeding_log_viz['Timestamp'] = pd.to_datetime(feeding_log_viz['Timestamp'])

# Plot
plt.figure(figsize=(12, 6))
plt.plot(feeding_log_viz['Timestamp'], feeding_log_viz['Nebraska Brand'], marker='o', color='green', label='Nebraska Brand')
plt.title('Feeding Trends for Nebraska Brand Over Time')
plt.xlabel('Date')
plt.ylabel('Amount Fed (lbs)')
plt.legend()
plt.grid(True)
plt.show()
```
